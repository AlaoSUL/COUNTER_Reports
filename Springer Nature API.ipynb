```python
#Import Libraries
import json
import requests
import csv
import pandas as pd
import lxml.etree as et
```

```python
###Springer Metadata API
#https://dev.springer.com/restfuloperations
#https://dev.springer.com/adding-constraints
#https://dev.springer.com/querystring-parameters
#https://dev.springer.com/example-metadata-response
springer_api_key = input("Please enter your springer api key: ") 

base_url_springer = 'http://api.springer.com/meta/v2/json'

# Building the Springer Metadata API parameters dictionary
url_params_springer = {}
url_params_springer["api_key"] = springer_api_key
url_params_springer["p"] = 100   #10 results will be returned


springer_keyword = '?q=type:Journal'
```

```python
springer = requests.get(base_url_springer + springer_keyword, params = url_params_springer).json()

springer
```

```python
df = pd.json_normalize(springer["records"])
df
```

```python
df.to_excel("./sampledf.xlsx")
```

# JATS Reponse for Meta API

JATS will return subjects fields.

```python
#Create empty lists to populate
book_title = [] #(book-title-group/book-title)
chapter_title = [] #(title-group/title)[]
book_doi = [] #(book-id)[0]
isbn_ppub = [] #(isbn)[0]
isbn_epub = [] #(isbn)[1]
primary_sub = [] #(meta_value)[1]
secondary_sub = [] #(meta_value)[2]
```

```python
d = et.parse("C:/Users/aclao89/Desktop/Online Usage Reports/SpringerNature/response_1663016734809.xml")  # file with your exact content
# e = d.getroot()
# e.xpath('.//book-title-group')
# e.xpath('.//book-title-group/book-title')[2].text # first p inside abstract


root=d.getroot()
```

```python
for title in root.iter("book-title"):
    print(title.text)
    book_title.append(title.text)

for chapter in root.iter("title"):
    print(chapter.text)
    chapter_title.append(chapter.text)    

for doi in root.iter("book-id"):
    print(doi.text)
    book_doi.append(doi.text)
    
    
for isbn in root.iter("isbn"):
    print(isbn.text)
    isbn_ppub.append(isbn.text)
    
for p_subject in root.iter("custom-meta"):
    print(p_subject[1].text)
    primary_sub.append(p_subject[1].text)

```

```python
sample_df = pd.DataFrame(list(zip(book_title, chapter_title, book_doi, isbn_ppub, primary_sub)), columns = ["Book", "Chapter", "DOI", "ISBN", "Primary Subject"])

sample_df.to_excel("./SpringerNature.sample_jats.xlsx")
```
